```python
nltk.download('stopwords')
from nltk.corpus import stopwords
import re
```

```python
engStopWords = set(stopwords.words('english'))
```

```python
engStopWords.remove('no')
engStopWords.remove('not')
```

```python
def sentenceFiltering(review):
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    wordList = review.split()
    wordList = [word for word in wordList if word not in engStopWords]
    return wordList
```

```python
def reviewSentences(review, tokenizer):
    sentenceList = []
    
    newSentences = tokenizer.tokenize(review.strip())
    
    for sentence in newSentences:
        if len(sentence) > 0:
            sentence = sentenceFiltering(sentence)
            sentenceList.append(sentence)
            
    return sentenceList
```

```python
from gensim.models import word2vec
```

```python
num_features = 300
min_word_count = 10
num_workers = 4
context = 10
```

```python
model = word2vec.Word2Vec(adaptedReviewList,\
                          workers=num_workers,\
                          vector_size=num_features,\
                          min_count=min_word_count,\
                          window=context)
```

```python
model.init_sims(replace=True)
```

```python
len(model.wv.vectors)
```

```python
model.wv.index_to_key
```

```python
model.wv.get_vector('dad')
```

```python
model.wv.most_similar('smart')
```

```python
index2wordSet = set(model.wv.index_to_key)
```

```python
def CreateSentenceVector(sentence, model, vectorDim):

    sentenceVec = np.zeros(vectorDim, dtype="float32")
    nwords = 0
    
    for word in sentence:
        if word in index2wordSet:
            nwords += 1
            sentenceVec = np.add(sentenceVec, model.wv.get_vector(word))
    
    if nwords == 0:
        return sentenceVec
    else:
        sentenceVec = np.divide(sentenceVec, nwords)
        return sentenceVec
```

```python
def CreateReviewVector(review, model, vectorDim):
    
    reviewLen = len(review)
    
    reviewVec = np.zeros(vectorDim)
    
    i = 0
    for i in range(reviewLen):
        vec = CreateSentenceVector(review[i], model, vectorDim)
        reviewVec = np.add(reviewVec, vec)
    
    reviewVec = np.divide(reviewVec, reviewLen)
    
    return reviewVec
```

```python
data['ReviewVector'] = data['ReviewSentenceList'].apply(lambda x: CreateReviewVector(x, model, 300))
```

# 3. Classification Models

```python
x, y = v2wData.drop(columns=['Sentiment2Cat', 'Sentiment3Cat', 'Sentiment4Cat']), v2wData.Sentiment2Cat
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtr, xtest, ytr, ytest = train_test_split(x, y, test_size=0.25, random_state=18)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

## 3. 1 Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB, ComplementNB, MultinomialNB
```

```python
naiveBayes = GaussianNB()
```

```python
naiveBayes.fit(xtr, ytr)
```

```python
predictionNB = naiveBayes.predict(xtest)
predictionNB
```

```python
print(f"Model's accuracy is: {round(accuracy_score(predictionNB, ytest), 4)*100}%")
print(f"Model's precision is: {round(precision_score(predictionNB, ytest, average='weighted'), 4)*100}%")
print(f"Model's recall is: {round(recall_score(predictionNB, ytest, average='weighted'), 4)*100}%")
print(f"Model's f1 is: {round(f1_score(predictionNB, ytest, average='weighted'), 4)*100}%")
```

```python
print(classification_report(predictionNB, ytest))
```
